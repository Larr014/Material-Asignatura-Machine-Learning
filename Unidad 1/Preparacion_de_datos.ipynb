### Convertir Fecha al mismo formato

```python
fechas = ["2024-01-18","2024/01/18","19-02-2024","02-19-2024","2024/05/20"]
fecha = pd.DataFrame(fechas,columns=['fecha'])
fecha
```

```python
import pandas as pd

def convertir_fecha(fecha):
  try:
    return pd.to_datetime(fecha, format='%Y-%m-%d')
  except:
    try:
      return pd.to_datetime(fecha, format='%Y-%d-%m')
    except:
      try:
        return pd.to_datetime(fecha, format='%d-%m-%Y')
      except:
        try:
          return pd.to_datetime(fecha, format='%m-%d-%Y')
        except:
          try:
            return pd.to_datetime(fecha, format='%Y/%m/%d')
          except:
            return None

fecha['fecha_convertida'] = fecha['fecha'].apply(convertir_fecha)
fecha
```

### Standard Scaler



*   Escala los datos para que tengan una media de 0 y una desviación estándar de 1.
*   Es útil cuando los datos siguen una distribución normal y se necesita que tengan una media de 0 y una desviación estándar de 1, como en algoritmos que asumen esta distribución (por ejemplo, regresión lineal, SVM).

*   Uso: Es útil cuando se necesita que los datos estén en un rango específico, como en redes neuronales donde las funciones de activación pueden ser sensibles a la escala de los datos.

## Correlacion de Pearson

Entre mayor valor, mayor relacion entre variables

```python
import pandas as pd

X = df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]
#y = df['species']
# Calcula la matriz de correlación de Pearson
correlation_matrix = X.corr(method='pearson')

# Imprime la matriz de correlación
print(correlation_matrix)

# Puedes establecer un umbral para seleccionar variables con alta correlación
threshold = 0.7  # Ejemplo de umbral

# Encuentra pares de variables con correlación absoluta por encima del umbral
high_corr_pairs = [(col1, col2) for col1 in correlation_matrix.columns
                    for col2 in correlation_matrix.columns
                    if col1 != col2 and abs(correlation_matrix[col1][col2]) > threshold]

# Imprime los pares de variables con alta correlación
print("Pares de variables con alta correlación:")
for pair in high_corr_pairs:
    print(pair)

# Puedes decidir eliminar una de las variables en cada par con alta correlación
# para reducir la redundancia en tus datos.
```

## Chi Square

Solo para variables Categoricas

```python
from sklearn.feature_selection import chi2

# Supongamos que 'y' es tu variable objetivo categórica y 'X' son tus variables predictoras categóricas
# Debes codificar las variables categóricas como numéricas antes de aplicar chi2
df9 = df.copy()
df9= df9.dropna()
df9['sex'] = df9['sex'].astype('category').cat.codes
df9['island'] = df9['island'].astype('category').cat.codes
df9['species'] = df9['species'].astype('category').cat.codes
X = df9[['sex', 'island']]
y = df9['species']
# Calcula el estadístico chi-cuadrado y los valores p
chi2_scores, p_values = chi2(X, y)

# Imprime los resultados
for i, feature in enumerate(X.columns):
    print(f"Característica: {feature}, Chi2 score: {chi2_scores[i]}, p-value: {p_values[i]}")

# Puedes establecer un umbral para los valores p para seleccionar las variables más relevantes
threshold = 0.05  # Ejemplo de umbral

# Selecciona las variables con valores p por debajo del umbral
selected_features = [feature for i, feature in enumerate(X.columns) if p_values[i] < threshold]

print("Variables seleccionadas:", selected_features)
```

## RFE

Pasos Eliminacion recursiva de caracteristicas

* Entrenamiento Inicial: Se entrena un modelo en el conjunto completo de características.
* Evaluación de Importancia: Se evalúa la importancia de cada característica utilizando el modelo entrenado.
* Eliminación de Características: Se eliminan las características menos importantes.
* Repetición: Se repiten los pasos anteriores en el conjunto reducido de características hasta alcanzar el número deseado de características.


```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier as RandomForest

# Supongamos que 'y' es tu variable objetivo categórica y 'X' son tus variables predictoras numéricas
df10 = df.copy()
df10 = df10.dropna()
X = df10[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = df10['species'].astype('category').cat.codes
modelo = RandomForest(random_state=0)
#modelo.fit(X,y)
rfe = RFE(estimator=modelo, n_features_to_select=3)

# Ajustar RFE al conjunto de datos
rfe.fit(X, y)

# Obtener las características seleccionadas
selected_features = rfe.support_
ranking = rfe.ranking_

print("Características seleccionadas:", selected_features)
print("Ranking de características:", ranking)
```
