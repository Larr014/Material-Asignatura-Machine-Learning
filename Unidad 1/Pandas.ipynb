#Transformar Datos

##Crear nueva variable

###Basandonos en una variable

Podemos asociar una funcion

```python
import random
def calcularGanancias(retweets):
  ganancias = retweets * random.randint(1,10)
  return ganancias

df["ganancias"] = df["retweets"].apply(calcularGanancias)
df
```

###Basandonos en una fila

axis=1 indica que ser√° aplicado a cada fila

```python
def popularidad(fila):
  resultado = fila["retweets"] / fila["followers"]
  return resultado

df["popularidad"] = df.apply(popularidad,axis=1)
df
```

#Agrupar

Consiste en unir filas que tengan un mismo valor dentro de una columna. Realizando una transformacion al resto de los datos

##Variable

```python
df_filtrado3 = df.drop(["full_text","user"],axis=1)
df_filtrado3
```

Asegurate de dejar solo las variables numericas

```python
df_filtrado3.groupby("country").mean()
```

##Variables Especificas

```python
agrupacion2 = df_filtrado3.groupby("country").agg({
    "followers":"sum",
    "retweets":"mean",
    "mentions":"max"
})
agrupacion2
```

##Multiple

```python
agrupado_multiple = df.groupby(["country","user"]).agg({
    "id":"count",
    "full_text":"count",
    "followers":"sum",
    "retweets":"mean",
    "mentions":"max"
}).sort_values("followers",ascending=False)
agrupado_multiple
```

Al realizar agrupacion es necesario considerar que generan nuevos indices y columnas. En el ejemplo country y user no serian columnas, si intentas df["user"] no funcionara. Para ello podemos realizar un reinicio de indexacion.

###Reinicio de indexacion

```python
desagrupado = agrupado_multiple.reset_index(1)
desagrupado
```

#Pivotar

Convertir cada valor unico de una columna en nuevas columnas

```python
pivote = desagrupado.pivot(columns="user")
pivote
```

```python
pivote_id = pivote["id"]
pivote_id
```

```python
pivote_id = pivote_id.fillna(0)
pivote_id
```
