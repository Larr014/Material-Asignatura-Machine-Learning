```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df1 = df.copy()
df1.drop_duplicates(inplace=True)
df1
```

```python
df1 = df1[['Survived','Pclass','Sex','Age','SibSp','Parch','Embarked']]
df1
```

```python
df1.isnull().sum()
```

```python
df1.dropna(subset=['Embarked'],inplace=True)
df1
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
df1
```

```python
df1.isnull().sum()
```

```python
df1
```

```python
df1['Sex'].unique()
```

```python
df1['Embarked'].unique()
```

```python
df1['Sex_id'] = df1['Sex'].astype('category').cat.codes
df1['Embarked_id'] = df1['Embarked'].astype('category').cat.codes
df1
```

```python
df1.to_csv('titanic_limpio.csv')
```

```python
df2 = df1[['Pclass','Sex_id','Age','SibSp','Parch','Embarked_id','Survived']]
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_cols = ['Pclass','Sex_id','Age','SibSp','Parch','Embarked_id']
df2[numerical_cols] = scaler.fit_transform(df2[numerical_cols])
df2
```

```python
df2.to_csv('titanic_final.csv')
```

```python
y = df2['Survived']
X = df2.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=4, stratify=y)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import ConfusionMatrixDisplay
def metricas(y_real,y_pred,labels):
  # Métricas de clasificación
  conf_matrix = confusion_matrix(y_test, y_pred)
  accuracy = accuracy_score(y_test, y_pred)

# Calculamos precision, recall y f1 para cada clase
  precision = precision_score(y_test, y_pred, average=None)
  precision_avg = precision_score(y_test, y_pred, average="macro")
  recall = recall_score(y_test, y_pred, average=None)
  recall_avg = recall_score(y_test, y_pred, average="macro")
  f1 = f1_score(y_test, y_pred, average=None)
  f1_avg = f1_score(y_test, y_pred, average="macro")
  print("Exactitud:", accuracy)
  print("Precisión por clase:", precision,'->', precision_avg)
  print("Sensibilidad por clase:", recall,'->', recall_avg)
  print("F1 Score por clase:", f1,'->', f1_avg)

  cm = confusion_matrix(y_test, y_pred)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm)
  disp.display_labels = labels
  disp.plot()
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
keras.utils.set_random_seed(4)#42
np.random.seed(4)#42
# Define the model
model1 = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(6,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(2, activation='softmax')  # 3 output classes for species
])

# Compile the model
#sparse_categorical_crossentropy si es es multiclase

model1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model1.save_weights('model1.weights.h5')

# Train the model
model1.fit(X_train, y_train, epochs=300, batch_size=64)
y_pred = model1.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
metricas(y_test,y_pred,['No Sobrevive','Si Sobrevive'])
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from tensorflow.keras import regularizers
keras.utils.set_random_seed(4)
np.random.seed(4)
# Define the model
model2 = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(6,),kernel_regularizer=regularizers.l2(0.02)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.02) ),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(2, activation='softmax')  # 3 output classes for species
])

# Compile the model
#sparse_categorical_crossentropy si es es multiclase

model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model2.load_weights('model1.weights.h5')

# Train the model
model2.fit(X_train, y_train, epochs=300, batch_size=64, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=4)])
y_pred = model2.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
metricas(y_test,y_pred,['No','Si'])
```
